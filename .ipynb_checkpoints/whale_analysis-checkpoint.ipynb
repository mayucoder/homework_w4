# Data Cleaning

In this section, you will need to read the CSV files into DataFrames and perform any necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.

Files:
1. whale_returns.csv
2. algo_returns.csv
3. sp500_history.csv

## Whale Returns

Read the Whale Portfolio daily returns and clean the data

```python
# Reading whale returns
whale_returns_csv = Path("./Resources/whale_returns.csv")
whale_df = pd.read_csv(whale_returns_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)
whale_df.sort_index(inplace=True)
whale_df.head()
```

```python
# Count nulls
whale_df.isnull().sum()
```

```python
# Drop nulls
whale_df = whale_df.dropna()
whale_df.isnull().sum()
```

```python
whale_df.head()
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data

```python
# Reading algorithmic returns
algo_returns_csv = Path("./Resources/algo_returns.csv")
algo_df = pd.read_csv(algo_returns_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)
algo_df.sort_index(inplace=True)
algo_df.head()
```

```python
# Count nulls
algo_df.isnull().sum()
```

```python
# Drop nulls
algo_df = algo_df.dropna()
algo_df.isnull().sum()
```

```python
algo_df.head()
```

## S&P 500 Returns

Read the S&P500 Historic Closing Prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P 500 Closing Prices
sp500_history_csv = Path("./Resources/sp500_history.csv")
sp500_df = pd.read_csv(sp500_history_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)
sp500_df.sort_index(inplace=True)
sp500_df.head()
```

```python
# Check Data Types
sp500_df.dtypes
```

```python
# Fix Data Types
sp500_df['Close'] = sp500_df['Close'].str.replace("$","").astype('float') # remove '$', convert string to float
sp500_df.dtypes
```

```python
# Calculate Daily Returns
#sp500_df.head()
sp500_daily_returns = sp500_df.pct_change()
sp500_daily_returns.head()
```

```python
# Drop nulls
sp500_daily_returns = sp500_daily_returns.dropna()
sp500_daily_returns.head()
```

```python
# Rename Column
sp500_daily_returns = sp500_daily_returns.rename(columns={"Close":"S&P 500"})
sp500_daily_returns.head()
```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
# Concatenate all DataFrames into a single DataFrame
combined_pf = pd.concat([whale_df, algo_df,sp500_daily_returns], axis="columns", join="inner")
combined_pf.sort_index(inplace=True)
combined_pf.head()
```

## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta. 

1. Plot the rolling standard deviation of the various portfolios along with the rolling standard deviation of the S&P 500 (consider a 21 day window). Does the risk increase for each of the portfolios at the same time risk increases in the S&P?
2. Construct a correlation table for the algorithmic, whale, and S&P 500 returns. Which returns most closely mimic the S&P?
3. Choose one portfolio and plot a rolling beta between that portfolio's returns and S&P 500 returns. Does the portfolio seem sensitive to movements in the S&P 500?
4. An alternative way to calculate a rolling window is to take the exponentially weighted moving average. This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the ewm with a 21 day half-life.

```python
# Calculate and plot the rolling standard deviation for
# the S&P 500 and whale portfolios using a 21 trading day window
combined_pf.rolling(window=21).std().plot(figsize= (25,10), title= '21 Day Rolling Standard Deviation')
```

```python
# Construct a correlation table
correlation = combined_pf.corr()
correlation
```

# INTERPRETATION OF CORRELATION TABLE
1. The movement of Soros fund and Algo 2 is closely related to the movement of S&P 500 wwhich means any change in S&P 500 will affect these two portfolios

```python
# Calculate 21 day rolling Beta for a single portfolio compared to the total market (S&P 500)
# (Your graph may differ, dependent upon which portfolio you are comparing)

rolling_covariance = combined_pf["SOROS FUND MANAGEMENT LLC"].rolling(window=21).cov(combined_pf["S&P 500"])
rolling_covariance
```

```python
rolling_variance = combined_pf['S&P 500'].rolling(window=21).var()
rolling_variance
```

```python
rolling_beta = rolling_covariance / rolling_variance
rolling_beta.plot(figsize=(20,10), title="Rolling 21-Day Beta of SOROS FUND MANAGEMENT LLC")
```

## INTERPRETAION OF BETA
rolling beta does not fall below zero and hence SOROS can be recommended as a portfolio to be considered

```python
# Calculate a 21 day rolling window using the exponentially weighted moving average (ewma). 
ewma = combined_pf.ewm(halflife= 21).std()
ewma
```

```python
ewma.plot(figsize= (20,10), title= "Exponentially Weighted Average")
```

## Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. (After all, if you could invest in one of two portfolios, each offered the same 10% return, yet one offered lower risk, you'd take that one, right?)

1. Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot.
2. Determine whether the algorithmic strategies outperform both the market (S&P 500) and the whales portfolios.

```python
# Calculate annualized Sharpe Ratios
sharpe_ratios = (combined_pf.mean() * 252) / (combined_pf.std() * np.sqrt(252))
sharpe_ratios
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratios.plot(kind="bar", title="Sharpe Ratios")
```

## Choose 3-5 custom stocks with at last 1 year's worth of historic prices and create a DataFrame of the closing prices and dates for each stock.

```python
# Read the first stock
tsla = Path("./Resources/TSLA.csv")
tsla_df = pd.read_csv(tsla, index_col="Date", parse_dates=True, infer_datetime_format=True)
tsla_df.sort_index(inplace=True)
tsla_df = tsla_df.rename(columns={"Close":"TSLA"})
tsla_df.head()
```

```python
# Read the second stock
fb = Path("./Resources/FB.csv")
fb_df = pd.read_csv(fb, index_col="Date", parse_dates=True, infer_datetime_format=True)
fb_df.sort_index(inplace=True)
fb_df= fb_df.rename(columns={"Close":"FB"})
fb_df.head()
```

```python
# Read the third stock
nflx = Path("./Resources/NFLX.csv")
nflx_df = pd.read_csv(nflx, index_col="Date", parse_dates=True, infer_datetime_format=True)
nflx_df.sort_index(inplace=True)
nflx_df = nflx_df.rename(columns={"Close":"NFLX"})
nflx_df.head()
```

```python
# Concatenate all stocks into a single DataFrame
new_pf = pd.concat([tsla_df, fb_df, nflx_df], axis= 'columns', join= 'inner')
new_pf.sort_index(inplace= True)
new_pf.head()
```

```python
#Calculate the returns for new portfolio
new_pf_returns= new_pf.pct_change()
new_pf_returns.head()
```

```python
# Pivot so that each column of prices represents a unique symbol
# NA
```

```python
# Drop Nulls
new_pf_returns= new_pf_returns.dropna()
new_pf_returns.head()
```

## Calculate the weighted returns for the portfolio assuming an equal number of shares for each stock

```python
# Calculate weighted portfolio returns
weights = [1/3, 1/3, 1/3]
new_pf_weighted_returns= new_pf_returns.dot(weights)
new_pf_weighted_returns.head()
```

## Join your portfolio returns to the DataFrame that contains all of the portfolio returns

```python
# Add your "Custom" portfolio to the larger dataframe of fund returns
final_pf= pd.concat([combined_pf, new_pf_weighted_returns], axis= 'columns', join= 'inner')
final_pf.tail()
```

```python

final_pf= final_pf.rename(columns= {0: 'Custom'})
final_pf.head()
```

```python
# Only compare dates where return data exists for all the stocks (drop NaNs)
final_pf.isnull().sum()
```

## Re-run the performance and risk analysis with your portfolio to see how it compares to the others

```python
# Risk
final_pf_cumulative= (1+ final_pf).cumprod()
final_pf_cumulative.plot(figsize= (20,10), title= 'Final portfolio cumulative returns')
```

```python
# Annualized Sharpe Ratios 252 days
final_pf_std= final_pf.std()*np.sqrt(252)
final_pf_std
```

```python
# Visualize the sharpe ratios as a bar plot
final_pf.rolling(window= 21).std().plot(figsize= (20,10), title= '21 Day final portfolio Rolling Std Deviation')
```

```python
# Create a correlation analysis
custom_sharpe_ratio= final_pf.mean()*252/final_pf_std
custom_sharpe_ratio.sort_values()
```

```python
# Beta
custom_sharpe_ratio.plot(kind= 'bar', title= 'CUSTOM SHARPE RATIOS')
```

```python
# Create a correlation analysis
correlation= final_pf.corr()
correlation
```

```python
# Covariance of Custom portfolio vs S&P 500
custom_covar= final_pf['Custom'].rolling(window= 21).cov(final_pf['S&P 500'])
custom_covar
```

```python
# Variance of S&P 500
sp_500variance= final_pf['S&P 500'].rolling(window=21).var()
sp_500variance
```

```python
# Beta of Custom portfolio
custom_beta= custom_covar/sp_500variance
custom_beta
```

```python
# Plotting Custom Beta
custom_beta.plot(figsize= (25,15), title= 'Custom Beta')
```
